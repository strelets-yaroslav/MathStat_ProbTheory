__Задача 1__

Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм. Используя уровень значимости 0.025 и правостороннюю критическую область, проверить данную гипотезу при условии, что в выборке из 100 шариков средний диаметр оказался равным 17.5 мм, а дисперсия известна и равна 4 мм.


Правосторонняя критическая область для уровня значимости 0.025 отвечает правилу двух сигм ($P(X < a + 2\sigma) = 0.975$). Т.е. требуется проверить, попадает ли средний диаметр в выборке в интервал $\left( a + 2\sqrt{\frac{4}{100}}, \infty \right) = \left( a + 0.4, \infty \right) = \left( 17.4, \infty \right)$. Видно, что средний диаметр выборки попал в этот интервал, следовательно, гипотеза не верна.

__Задача 2__

Известно, что генеральная совокупность распределена нормально с известным средним квадратическим отклонением 16. Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95, если выборочное среднее равно 80, а объём равен 256.


Надёжность 0.95 и нормальное распределение совокупности означает, что для оценки применяется правило двух сигм. Т.е. доверительный интервал будет выглядеть следующим образом: $\left( a - 2\frac{16}{\sqrt{256}}, a + 2\frac{16}{\sqrt{256}} \right) = \left( 80 - 2, 80 + 2 \right) = (78, 82)$.

__Задача 3__

Продавец утверждает, что средний вес пачки печенья составляет 200 г. Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет:
```
202, 203, 199, 197, 195, 201, 200, 204, 194, 190
```
Известно, что их веса распределены нормально. Верно ли утверждение продавца, если учитывать, что уровень значимости равен 1%?


Доверительный интервал, учитывая, что среднее квадратическое отклонение нам неизвестно, а выборка небольшая, будем определять, используя t-распределение: $P \left( 
\overline{X} - t_{(1 - 0.01/2, \: 9)} \cdot \dfrac{\sigma_X}{\sqrt{10}} 
\leq M(X) \leq
\overline{X} + t_{(1 - 0.01/2, \: 9)} \cdot \dfrac{\sigma_X}{\sqrt{10}} 
\right) = 1 - 0.01$. По таблице квантилей t-распределения $t_{(1 - 0.01/2, \: 9)} = 3.2498$.

````python
import numpy as np
````

````python
samples = np.array([202, 203, 199, 197, 195, 201, 200, 204, 194, 190])
````

````python
n = samples.shape[0]
m = samples.mean()
s = samples.std(ddof=1)
print(m, s, n)
````

````python
t = 3.2498
````

````python
(m - t * s / np.sqrt(n), m + t * s / np.sqrt(n))
````

Значение изначального утверждения (200г) попадает в этот интервал, следовательно, гипотеза верна.
